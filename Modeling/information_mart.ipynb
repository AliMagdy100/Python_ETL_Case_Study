```python
import pandas as pd
```

```python
orders = pd.read_csv('../staging_2/orders/orders.csv')
orders.head(3)
```

```python
order_items = pd.read_csv('../staging_2/order_items/order_items.csv')
order_items.head(3)
```

```python
products = pd.read_csv('../staging_2/products/products.csv')
products.head(3)
```

```python
# Merge orders with order_items on 'order_id' using left join
merged_orders_items = pd.merge(orders, order_items, on='order_id', how='inner')

# Merge the result with products on 'product_id' using left join
final_merged_df = pd.merge(merged_orders_items, products, on='product_id', how='inner')

# Display the final merged DataFrame
final_merged_df.head(3)
```

```python
final_merged_df.to_csv('../inforamtion_mart/orders_products_details/orders_products_details.csv' , index= False)
```
