```python
import pandas as pd
```

```python
customers = pd.read_csv('../staging_1/customers/customers.csv')
len(customers)
```

```python
stores = pd.read_csv('../staging_1/stores/stores.csv')
stores.head()
```

```python
merged_df = pd.merge(customers, stores[['state', 'city','store_id']], on=['state', 'city'], how='left')
merged_df
```

```python
merged_df['local'] =~ merged_df['store_id'].isna()
merged_df['local'].value_counts()
```

```python
merged_df.to_csv('../staging_2/customers/customers.csv' , index= False)
```
